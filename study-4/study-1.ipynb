```python
import pandas as pd
import numpy as np
```

```python
#Загрузка датасета
df = pd.read_csv('vgsales_10.csv')
#Просмотр и исследование датасета
df.info()
df.shape
df
```

```python
#Обработка датасета

#Удаление хотя бы одного признака
df.drop(columns=['Other_Sales'])
```

```python
#Замена значений
df = df.replace({'Platform' : 'PSP'}, 'psp')
df
```

```python
# Использование простейшего подстановщика
df = pd.DataFrame(
    [['PS3', 'Adventure', 1],
    ['GC', 'Sports',],
    ['SAT', 'Strategy',]],
    columns=['Platform', 'Genre', 'INT'])
df
```

```python
#Выбрать один признак в качестве целевой переменной и обосновать данный выбор
from sklearn.impute import SimpleImputer

simp = SimpleImputer(missing_values=np.nan)
simp = simp.fit(df[['INT']])

df['INT'] = simp.transform(df[['INT']])
df.isnull().mean()
#Число
```

```python
#Написать заключение по проведенному исследованию
# Было интересно

#Сохранить датасет
df.to_csv('Ogurcov.csv', index=False)
```
